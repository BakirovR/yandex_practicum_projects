### Поиск и исправление ошибок в данных

Проверим: есть ли заказы, которым соответствуют разные клиенты.

```python
report = df.groupby('order_id').agg({'customer_id': 'nunique'}).query(
    'customer_id >=2').sort_values(by='customer_id', ascending=False).reset_index()
report
```

Таким образом, существуют одинаковые наименования заказов, которым соответствуют различные клиенты, что неправильно. Посмотрим на часть заказов.

```python
a=report['order_id'][0:2]
df.query('order_id in @a')
```

Возможно, есть проблема с идентификацией клиента при повторном входе. Сообщим данную информацию ответственным коллегам. Ввиду небольшого кол-ва подобных аномалий осуществим их удаление.

```python
df = df.loc[~df['order_id'].isin(report['order_id'])]
report = df.groupby('order_id').agg({'customer_id': 'nunique'}).query(
    'customer_id >=2').sort_values(by='customer_id', ascending=False).reset_index()
report.count()
```

Удаление выполнено успешно.

Проверим: есть ли заказы, которым соответствуют разные даты.

```python
report = df.groupby('order_id').agg({'date': 'nunique'}).query(
    'date >=2').sort_values(by='date', ascending=False).reset_index()
report
```

Таким образом, существуют одинаковые наименования заказов, которым соответствуют различные даты, что неправильно. Посмотрим на часть заказов.

```python
a=report['order_id'][:4]
df.query('order_id in @a')
```

Возможно, произошёл технический сбой.

Сообщим данную информацию ответственным коллегам. 

Осуществим удаление данных аномалий.

```python
df = df.loc[~df['order_id'].isin(report['order_id'])]
report = df.groupby('order_id').agg({'date': 'nunique'}).query(
    'date >=2').sort_values(by='date', ascending=False).reset_index()
report.count()
```

Удаление выполнено успешно.

Из документации к библиотеке `pymystem3` следует, что она хорошо работает с целиковым текстом. Поэтому объединим значения в столбце `product` в единый текст с разделителем ввиде '$', а затем проведем обратную операцию с помощью функции `split`.

### Нахождение самых прибыльных и продаваемых товаров внутри категории

Найдём самые прибыльные товары внутри категорий. Напишем функцию top5 и осуществим её применение в цикле.

```python
def top5(category, sort_parameter):
    report = df.query('category ==@category').groupby('product')\
                               .agg({'category':'first', 'order_id':'nunique', 'customer_id':'nunique', 
                                     'quantity':'sum', 'price':'first', 'total_price':'sum'})\
                               .reset_index().sort_values(by=sort_parameter, ascending=False)
    report.columns = ['product', 'category', 'orders', 'customers', 'quantity', 'price','total_price']

    return report.head()
```

```python
# применение функции в цикле
for category in df['category'].unique():
    print()
    display(f"ТОП 5 самых прибыльных товаров в категории: «{category}»")
    display(top5(category, 'total_price'))
    print()
```

Заметим, что:

  - самыми прибыльными в категории `дом` являются: простыни и сумки-тележки;
  
  
  - самыми прибыльными в категории `кухня` являются: коврики, скатерти, чехлы для гладильной доски и одежды;
  
  
  - самыми прибыльными в категории `дача` являются: урны, веники и корыта;
  
  
  - самыми прибыльными в категории `ремонт и строительство` являются: лестницы и крепеж.
  



Найдем самые продаваемые товары. Снова применим функцию топ5 в цикле.

```python
# применение функции в цикле
for category in df['category'].unique():
    print()
    display(f"ТОП 5 самых продаваемых товаров в категории: «{category}»")
    display(top5(category, 'order_id'))
    print()
```

Для определения количества заказов и дохода внутри категорий по сезонам, месяцам, дням недели и часам напишем функцию `category_analysys` и осуществим её вызов в цикле.

```python
# создание функции category_analysys
def category_analysys(category, period):
    report = df.query('category ==@category')
    report = report.groupby([period]).agg(
        {'total_price': 'sum', 'order_id': 'nunique'}).reset_index()
    report.columns = [period,'total_price', 'number_of_orders']
    return report
```

## Проверка статических гипотез

Нам необходимо проверить две гипотезы:

- доход с заказов, состоящих из нескольких категорий, и с заказов только с одной категорией различаются;
- доход с заказов в различные сезоны различается.



Никакие экспериментально полученные данные не подтвердят гипотезу. Это фундаментальное ограничение. При условии, что гипотеза верна, данные могут лишь не противоречить ей или, наоборот, показывать очень маловероятные результаты.

Но и в том, и в другом случае нет оснований утверждать, что выдвинутая гипотеза доказана.

Итак, если данные гипотезе не противоречат, тогда мы её не отвергаем. Если же мы приходии к выводу, что вероятность получить такие данные очень низка, появляется основание отбросить эту гипотезу. Задача статистического теста как раз в том, чтобы эту вероятность найти.

Статистическая гипотеза — на то и статистическая, чтобы утверждать что-то о параметре распределения всей генеральной совокупности исследуемых объектов.

Проверка гипотез начинается с формулировки нулевой гипотезы H₀ (от англ. hypothesis — «гипотеза»).

Исходя из H₀ формулируется альтернативная гипотеза H₁. Это утверждение, которое принимается верным, если отбрасывается H₀.

Когда мы сформулировали нулевую и альтернативную гипотезу, остается только на основе данных сделать выбор: отклонить нулевую гипотезу в пользу альтернативной или нет.


Для этого нужно посчитать вероятность получить наблюдаемое на выборке значение (чаще всего, среднее) при условии, что нулевая гипотеза верна. Если вероятность большая — отвергать нулевую гипотезу вряд ли стоит.

Перейдем к проверке гипотез.


###  Доход с заказов, состоящих из нескольких категорий, и с заказов только с одной категорией различаются

Проверим гипотезу о том, что средний чек заказов, состоящих из нескольких категорий, и заказы только с одной категорией различаются.

  - Нулевая гипотеза: доход с заказов, состоящих из нескольких категорий, и заказов с одной категорией не отличается;
  
  - Альтернативная гипотеза: доход с заказов, состоящих из нескольких категорий, и заказов с одной категорией различается.
  
  
Для проверки гипотезы критический уровень статистической значимости примем  0.05.

Подготовим значения для проверки. Создадим `sample1` с одной категорией и `sample2` с несколькими.


```python
# группировка заказов и подсчёт категорий
report = df.groupby(['order_id']).agg({'total_price': 'sum', 'category': 'nunique'}).reset_index()
report.head()

```

```python
# разбивка по кол-ву категорий
sample1 = report.query('category ==1')['total_price']
sample2=report.query('category !=1')['total_price']
```

Посмотрим на разницу между средними значениями.

```python
sample2.mean()/sample1.mean()
```

Разница более чем в полтора раза. Проверим гипотезу.

```python

alpha = 0.05 # критический уровень статистической значимости 
results = stats.mannwhitneyu(sample1, sample2)[1]

print('p-значение:', results)

if (results < alpha):
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу")
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно даже меньше 1%, поэтому делаем вывод о значимом различии доходов с заказов, состоящих из нескольких категорий, и заказов с одной категорией.

###  Доход с заказов в различные сезоны различается.

Сформируем датафрейм с группировкой по сезонам и заказам.

```python
report = df.groupby(['season', 'order_id'])['total_price'].sum().reset_index()
report.head()
```

Выделим доход по сезонам.

```python
winter_sample = report.query('season =="Зима"')['total_price']
spring_sample = report.query('season =="Весна"')['total_price']
summer_sample = report.query('season =="Лето"')['total_price']
autumn_sample = report.query('season =="Осень"')['total_price']
```

Создадим функцию `stat_check` и последовательно сравним доходы с заказов в различные сезоны. Применим поправку Бонферрони для 6 тестов.

```python
def stat_check(sample1, sample2):
    print(f'Разница между средними значениями : {sample1.mean() / sample2.mean()}')
    
    alpha = 0.05/6 # критический уровень статистической значимости 
    results = stats.mannwhitneyu(sample1, sample2)[1]

    print('p-значение:', results)

    if (results < alpha):
        print("Отвергаем нулевую гипотезу")
    else:
        print("Не получилось отвергнуть нулевую гипотезу")
```

Сравним зиму и весну.

  - Нулевая гипотеза: доход зимой и весной не отличается;
  - Альтернативная гипотеза: доход зимой и весной отличается.

```python
stat_check(winter_sample,spring_sample)
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно составляет 78%. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между доходами зимой и весной.

Сравним зиму и лето.

  - Нулевая гипотеза: доход зимой и летом не отличается;
  - Альтернативная гипотеза: доход зимой и летом отличается.

```python
stat_check(winter_sample,summer_sample)
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно составляет 62%. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между доходами зимой и летом.
