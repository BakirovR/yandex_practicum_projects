**Описание проекта**

Мы работаем в интернет-магазине «Стримчик», который продаёт по всему миру компьютерные игры. Из открытых источников доступны исторические данные о продажах игр, оценки пользователей и экспертов, жанры и платформы (например, Xbox или PlayStation). 


Перед нами данные до 2016 года. Представим, что сейчас декабрь 2016 г., и мы планируем кампанию на 2017-й. Нужно отработать принцип работы с данными. Неважно, прогнозируем ли мы продажи на 2017 год по данным 2016-го или же 2027-й — по данным 2026 года.


В наборе данных попадается аббревиатура ESRB (Entertainment Software Rating Board) — это ассоциация, определяющая возрастной рейтинг компьютерных игр. ESRB оценивает игровой контент и присваивает ему подходящую возрастную категорию, например, «Для взрослых», «Для детей младшего возраста» или «Для подростков».


**Цель исследования** —   выявить определяющие успешность игры закономерности. Это позволит сделать ставку на потенциально популярный продукт и спланировать рекламные кампании.

Предварительно необходимо осуществить предобработку данных, посчитать и добавить в таблицу суммарные продажи во всех регионах.


**Ход исследования**

### Пропуски значений
Сначала посчитаем, сколько в таблице пропущенных значений. Для этого достаточно двух методов `pandas`:

```python
# подсчёт пропусков

df.isna().sum()
```

Таким образом мы обнаружили пропущенные значения в шести  из одиннадцати столбцов.

Проверим, какую долю составляют пропущенные значения в каждом из столбцов с пропусками. Для этого будем использовать цикл и полученные значения выведем на экран.

```python
# доля пропущенных значений в %

for name in df.columns:
    if df[name].isna().sum() > 0:

        display(
            f' {name.ljust(25)} {df[name].isna().sum() / (df[name].isna().sum() + df[name].count()): .2%}')
```

Заметим, что наряду с большими значениями в столбцах `critic_score`,`user_score  `,`rating` ,существуют незначительные пропуски в  `name`,`year_of_release`,`genre`.

Рассмотрим столбцы с пропусками по отдельности.

Т.к. процент пустых значений в стобцаx `name, genre` < 5% осуществим их удаление.

```python
# удаление строк с пропусками в столбцах, перечисленных в списке subset
df = df.dropna(subset=['name', 'genre'])
```

Осуществим проверку.

```python
df.isna().sum()
```

Пустые строки удалены.

Рассмотрим пустые значения в столбце `year_of_release`.

```python
# вывод строк с пустыми значениями в столбце year_of_release
df[df['year_of_release'].isna()].head()
```

Заметим, что для части значений можно вытащить год выпуска из названия игры. К примеру, из строки под индексом 183 можно взять 2004 год выпуска.

```python
# присвоение пустым строкам столбца year_of_release года из названия игры
df.loc[df['year_of_release'].isna(),'year_of_release'] = df[df['year_of_release'].isna()]['name'].str.extract('(\d{4})')[0] 
```

```python
# пример к комментарию ревьюера
df.iloc[[333,338,617]]
```

Осуществим проверку.

```python
df.iloc[[183,377,475]]  
```

```python
# подсчёт оставшихся пустых значений в столбце year_of_release
df['year_of_release'].isna().sum()
```

Значения заменены. Но большую часть не удалось заполнить, поэтому, т.к. их меньше 5% осуществим удаление этих строчек.

```python
# удаление строк с пропусками в столбцах, перечисленных в списке subset
df = df.dropna(subset=['year_of_release'])
```

Выполним проверку.

```python
df['year_of_release'].isna().sum()
```

Рассмотрим столбцы `critic_score, user_score, rating`.

```python
# вывод таблицы
df.head()
```

```python
df['user_score'].value_counts()
```

Возможно, отсутствие значений в critic_score, user_score, rating связано с тем, что на момент выхода игр в прокат система оценок не была развита или развитие интернета не позволяло осуществить действия по оценке игр.

Пропуски в `critic_score, user_score` заполнять некорректно - любое заполнение исказит результаты корреляционного анализа и стат-тестов. Пропуски в `rating` заполним заглушкой (например, значением "undefined"), потому что это категориальная переменная. Также заметим, что в столбце `user_score` большое кол-во значений tbd - сокращение от To Be Determined. Служит для отметки неясностей или пропусков, которые надо заполнить. По сути tbd означает, что для формирования рейтинга не хватает обзоров. Чтобы сохранить возможность работы со столбцом(корреляционный анализ, стат-тесты) заменим tbd на NaN. 

```python
# замена tbd на NaN 
df.loc[df['user_score'] == 'tbd' , 'user_score'] = np.nan
```

```python
# замена NaN на 'undefined' в столбце rating
df['rating'].fillna('undefined', inplace=True)
```

Осуществим проверку.

```python
df.loc[df['user_score'] == 'tbd' , 'user_score'].count()
```

```python
df.head()
```

Таким образом, замены осуществлены корректно.

### Определение самых популярных платформ по регионам

Для определения самых популярных платформ по регионам создадим функции `district_top5` и `district_pie`.

```python
# функция формирования таблицы региона
def district_top5(column, district):
    return report_actual.pivot_table(index=column, values=district, aggfunc='sum').sort_values(by=district, ascending=False).head()
```

```python
# функции построения круговых диаграмм для регионов
def district_pie(check_param,district_name,district_label):
    plt.figure(figsize=(10,10))
    plt.suptitle("Распредение по регионам" )
    for i, name in enumerate(district_name):
        plt.subplot(2,2,i+1)
        report = district_top5(check_param, name)
        colors = sns.color_palette('pastel')[0:5]
        plt.pie(report[name],labels=report.index, autopct='%.0f%%',colors=colors)
        plt.title(district_label[i])
```

```python
# создание списков регионов и названий для подписей к диаграммам 
district_name = list(report_actual.columns[4:8])
district_label = ['Продажи в Северной Америке','Продажи в Европе','Продажи в Японии','Продажи в остальном мире']
```

Применим функцию district_pie к списку регионов.

```python
# использование функции district_pie для построения круговых диаграмм
district_pie('platform', district_name, district_label)
```

Также оценим продажи по платформам в регионах, используя функцию district_top5.

```python
for i in range(len(district_name)):
    print(district_top5('platform', district_name[i]))
    
```

## Проверка гипотез

Нам необходимо проверить две гипотезы:

- Средние пользовательские рейтинги платформ Xbox One и PC одинаковые;

- Средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

Никакие экспериментально полученные данные не подтвердят гипотезу. Это фундаментальное ограничение. При условии, что гипотеза верна, данные могут лишь не противоречить ей или, наоборот, показывать очень маловероятные результаты.

Но и в том, и в другом случае нет оснований утверждать, что выдвинутая гипотеза доказана.

Итак, если данные гипотезе не противоречат, тогда мы её не отвергаем. Если же мы приходии к выводу, что вероятность получить такие данные очень низка, появляется основание отбросить эту гипотезу. Задача статистического теста как раз в том, чтобы эту вероятность найти.

Статистическая гипотеза — на то и статистическая, чтобы утверждать что-то о параметре распределения всей генеральной совокупности исследуемых объектов.

Проверка гипотез начинается с формулировки нулевой гипотезы H₀ (от англ. hypothesis — «гипотеза»).

Исходя из H₀ формулируется альтернативная гипотеза H₁. Это утверждение, которое принимается верным, если отбрасывается H₀.

Когда мы сформулировали нулевую и альтернативную гипотезу, остается только на основе данных сделать выбор: отклонить нулевую гипотезу в пользу альтернативной или нет.


Для этого нужно посчитать вероятность получить наблюдаемое на выборке значение (чаще всего, среднее) при условии, что нулевая гипотеза верна. Если вероятность большая — отвергать нулевую гипотезу вряд ли стоит.

В Python есть специальный метод для проверки гипотезы о равенстве среднего двух генеральных совокупностей по взятым из них выборкам: scipy.stats.ttest_ind (array1, array2,equal_var).
Методу передают параметры:
array1, array2 — массивы, содержащие выборки;
equal_var (от англ. equal variance — «равная дисперсия») — необязательный параметр, задающий, считать ли равными дисперсии выборок. Передаётся как equal_var = True или equal_var = False (True — считать, False — не считать).

Перейдем к проверке гипотез.

### Проверка первой гипотезы

Итак, возьмем столбец `user_score`, разделим его на части по платформам  Xbox One и PC, найдем для этих частей средние, сравним их, а потом проведем стат-тест, чтобы понять есть ли реальное различие между средними. Пустые значения в столбце `user_score` учитывать не будем.

```python
# выбор столбца user_score платформы Xbox One и определение среднего
report = report_actual.query('platform == "XOne"')
sample_1 = report['user_score'].dropna()
sample_1.mean()
```

```python
# выбор столбца user_score платформы PC и определение среднего
report = report_actual.query('platform == "PC"')
sample_2 = report['user_score'].dropna()
sample_2.mean()
```

Средние Xbox One и PC близки друг к другу.

Проведем стат-тест.

Сформулируем нулевую и альтернативную гипотезы.

Нулевая гипотеза: средние пользовательские рейтинги платформ Xbox One и PC одинаковые.

Альтернативная гипотеза: средние пользовательские рейтинги платформ Xbox One и PC разные.

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем нулевую гипотезу
```

```python
results = st.ttest_ind(
    sample_1, 
    sample_2)
```

```python
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно почти 16%.  Средние Xbox One и PC близки друг к другу. Это явно слишком большая вероятность, чтобы делать вывод о значимом различии между средними  Xbox One и PC.


### Проверка второй гипотезы

Итак, возьмем столбец `user_score`, разделим его на части по жанрам  Action  и Sports, найдем для этих частей средние, сравним их, а потом проведем стат-тест, чтобы понять есть ли реальное различие между средними. Пустые значения в столбце `user_score` учитывать не будем.

```python
# выбор столбца user_score для жанра Action  и определение среднего
report = report_actual.query('genre == "Action"')
sample_1 = report['user_score'].dropna()
sample_1.mean()
```

```python
# выбор столбца user_score для жанра Sports  и определение среднего
report = report_actual.query('genre == "Sports"')
sample_2 = report['user_score'].dropna()
sample_2.mean()
```

```python
sample_1.mean()/sample_2.mean()
```

Среднее жанра Action в 1.3 больше жанра Sports.

Проведем стат-тест.

Сформулируем нулевую и альтернативную гипотезы.

Нулевая гипотеза: средние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») одинаковые.

Альтернативная гипотеза: cредние пользовательские рейтинги жанров Action (англ. «действие», экшен-игры) и Sports (англ. «спортивные соревнования») разные.

```python
alpha = .05 # критический уровень статистической значимости
# если p-value окажется меньше него - отвергнем нулевую гипотезу
```

```python
results = st.ttest_ind(
    sample_1, 
    sample_2)
```

```python
print('p-значение: ', results.pvalue)

if results.pvalue < alpha:
    print("Отвергаем нулевую гипотезу")
else:
    print("Не получилось отвергнуть нулевую гипотезу") 
```

Полученное значение p-value говорит о том, что вероятность получить такое или большее различие случайно даже меньше 1%,  средняя оценка пользователей из разных жанров неодинакова. Поэтому, делаем вывод о значимом различии между cредними пользовательскими рейтингами жанров Action и Sports.
